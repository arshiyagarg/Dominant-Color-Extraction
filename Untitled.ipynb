```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
img = cv2.imread('color_extraction.jpeg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
img = cv2.resize(img,(353,530))
```

```python
img.shape
```

```python
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
X = img.reshape((-1,3))
```

```python
X.shape
```

```python
from sklearn.cluster import KMeans
```

```python
k = 1
model = KMeans(n_clusters=k)
```

```python
model.fit(X)
```

```python
centroids = model.cluster_centers_
```

```python
print(centroids)
```

```python
colors = np.array(centroids,dtype='uint8')
```

```python
print(colors)
```

```python
#MAKING COLOR PATCH
cnt = 1
for color in colors:
    plt.subplot(1,k,cnt)
    cnt += 1
    plt.axis('off')
    mat = np.zeros((100,100,3),dtype='uint8')
    mat[:,:,:] = color
    plt.imshow(mat)
    
plt.show()    
```
